```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_EXPERIMENTS = 7
N_CELLS = 8

df_ls = []
for i in range(N_EXPERIMENTS):
    target_df = pd.read_excel('0{}-12022021.xlsx'.format(i + 1), sheet_name='ForceCells', usecols=[i for i in range(N_CELLS * 3)])

    df_ls.append(target_df)

target_df = pd.concat(df_ls, axis=0)
target_df
```

```python
plt.figure(figsize=(40,40))
for i in range (1, N_CELLS + 1):
    plt.subplot(4, 2, i)
    plt.plot(target_df['Time'], target_df['F{}x'.format(str(i))], label='X')
    plt.plot(target_df['Time'], target_df['F{}y'.format(str(i))], label='Y')
    plt.plot(target_df['Time'], target_df['F{}z'.format(str(i))], label='Z')
    plt.legend()
    plt.title('Cell {}'.format(str(i)))
    plt.xlabel('Time (s)')
    plt.ylabel('Force (N/m)')

plt.show()
```

```python
def rotate_vector(v, axis, angle):
    if axis == 0:
        # X
        v = v.dot(np.array([[1, 0, 0], [0, np.cos(np.radians(angle)), np.sin(np.radians(angle))], [0, np.sin(np.radians(angle)), np.cos(np.radians(angle))]]))
    elif axis == 1:
        # Y
        v = v.dot(np.array([[np.cos(np.radians(angle)), 0, np.sin(np.radians(angle))], [0, 1, 0], [-np.sin(np.radians(angle)), 0, np.cos(np.radians(angle))]]))
    elif axis == 2:
        # Z
        v = v.dot(np.array([[np.cos(np.radians(angle)), -np.sin(np.radians(angle)), 0], [np.sin(np.radians(angle)), np.cos(np.radians(angle)), 0], [0, 0, 1]]))
    else:
        raise ValueError('Invalid axis')

    return pd.Series(v)
    
rotations = {
    1: [180, 90, 0],
    2: [180, 90, 0],
    3: [180, 0, -90],
    4: [0, 0, -90],
    5: [0, 0, 0],
    6: [0, 180, 0],
    7: [0, 90, 0],
    8: [0, 0, 90],
}

for i in range (1, N_CELLS + 1):
    cols = ['F{}x'.format(str(i)), 'F{}y'.format(str(i)), 'F{}z'.format(str(i))]
    for ax in range(3):
        target_df[cols] = target_df[cols].apply(lambda v: rotate_vector(v, ax, rotations[i][ax]), axis=1)


# plt.figure(figsize=(40,40))
# for i in range (1, N_CELLS + 1):
#     plt.subplot(4, 2, i)
#     plt.plot(target_df['Time'], target_df['F{}x'.format(str(i))], label='X')
#     plt.plot(target_df['Time'], target_df['F{}y'.format(str(i))], label='Y')
#     plt.plot(target_df['Time'], target_df['F{}z'.format(str(i))], label='Z')
#     plt.legend()
#     plt.title('Cell {}'.format(str(i)))
#     plt.xlabel('Time (s)')
#     plt.ylabel('Force (N/m)')

# plt.show()
```

```python
df_ls = []
for i in range(N_EXPERIMENTS):
    features_df = pd.read_excel('0{}-12022021.xlsx'.format(i + 1), sheet_name='H3processed')

    df_ls.append(features_df)

features_df = pd.concat(df_ls, axis=0)
features_df
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.multioutput import MultiOutputRegressor
```

```python
X = []
Y = []
for i in range(N_CELLS):
    X.append(target_df['F{}z'.format(i + 1)].values)
    Y.append(target_df[['F{}x'.format(i + 1), 'F{}y'.format(i + 1)]].values)

X = np.concatenate(X, axis=0)[:, np.newaxis]
Y = np.concatenate(Y, axis=0)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
```

```python
scaler = StandardScaler().fit(X_train)

X_train_norm = scaler.transform(X_train)
X_test_norm =  scaler.transform(X_test)
```

```python
param_grid = {}
#     'n_estimators': [10, 20, 50, 100, 500],
#     'criterion': ['mse', 'mae'],
#     'max_depth': [1, 2, 5, 10],

# }

rf = RandomForestRegressor(random_state=0)
clf = MultiOutputRegressor(rf, n_jobs=-1)

clf = GridSearchCV(clf, param_grid, cv=5, scoring='neg_mean_squared_error')
clf.fit(X_train_norm, Y_train)

print('Best params: {}'.format(clf.best_params_))


test_preds = clf.best_estimator_.predict(X_test_norm)

print('Best train MAE: {}'.format(mean_absolute_error(Y_train, clf.best_estimator_.predict(X_train_norm), multioutput='raw_values')))
print('Best train MSE: {}'.format(mean_squared_error(Y_train, clf.best_estimator_.predict(X_train_norm), multioutput='raw_values')))
print('Best train R2: {}'.format(r2_score(Y_train, clf.best_estimator_.predict(X_train_norm), multioutput='raw_values')))

print('Best test MAE: {}'.format(mean_absolute_error(Y_test, test_preds, multioutput='raw_values')))
print('Best test MSE: {}'.format(mean_squared_error(Y_test, test_preds, multioutput='raw_values')))
print('Best test R2: {}'.format(r2_score(Y_test, test_preds, multioutput='raw_values')))
```
